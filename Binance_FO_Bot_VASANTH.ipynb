```python

#  INSTALL REQUIRED LIBRARIES

!pip install requests


#  IMPORTS

import time
import hmac
import hashlib
import requests
import json
import argparse
import logging
from getpass import getpass
from datetime import datetime


#  CONFIGURATION

BASE_URL = "https://testnet.binancefuture.com"

# Secure API Input
API_KEY = getpass("Enter your Binance Futures Testnet API key: ")
API_SECRET = getpass("Enter your Binance Futures Testnet Secret key: ")


#  LOGGING SETUP

logging.basicConfig(
    filename='trading_bot.log',
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)


#  SIGN REQUESTS

def sign_params(params):
    query = '&'.join([f"{k}={v}" for k, v in params.items()])
    signature = hmac.new(API_SECRET.encode(), query.encode(), hashlib.sha256).hexdigest()
    params['signature'] = signature
    return params


#  REST HELPERS

def send_request(http_method, endpoint, params=None):
    url = BASE_URL + endpoint
    headers = {"X-MBX-APIKEY": API_KEY}
    try:
        if http_method == "GET":
            resp = requests.get(url, headers=headers, params=params)
        elif http_method == "POST":
            resp = requests.post(url, headers=headers, params=params)
        else:
            raise ValueError("Unsupported HTTP method.")

        resp.raise_for_status()
        logging.info(f"{http_method} {endpoint} ✅ {resp.status_code}")
        return resp.json()

    except Exception as e:
        logging.error(f"{http_method} {endpoint}  {e} -> {getattr(resp, 'text', '')}")
        return {"error": str(e), "response": getattr(resp, 'text', '')}


#  CORE FUNCTIONS


def get_server_time():
    """Get Binance server time"""
    return send_request("GET", "/fapi/v1/time")

def get_account_info():
    """Fetch account info"""
    params = {"timestamp": int(time.time() * 1000)}
    signed = sign_params(params)
    return send_request("GET", "/fapi/v2/account", signed)

def place_order(symbol, side, order_type, quantity, price=None):
    """Place a Market or Limit order"""
    params = {
        "symbol": symbol.upper(),
        "side": side.upper(),
        "type": order_type.upper(),
        "quantity": quantity,
        "timestamp": int(time.time() * 1000),
        "recvWindow": 5000
    }
    if order_type.upper() == "LIMIT":
        if price is None:
            raise ValueError("Price is required for LIMIT orders.")
        params["price"] = price
        params["timeInForce"] = "GTC"

    signed = sign_params(params)
    result = send_request("POST", "/fapi/v1/order", signed)

    if "error" not in result:
        print("\n Order placed successfully!")
        print(json.dumps(result, indent=2))
    else:
        print("\n Failed to place order.")
        print(json.dumps(result, indent=2))

    return result


#  CLI INTERFACE

def main():
    parser = argparse.ArgumentParser(description="Simplified Binance Futures Testnet Trading Bot")

    parser.add_argument("--symbol", required=True, help="Trading pair symbol (e.g., BTCUSDT)")
    parser.add_argument("--side", required=True, choices=["BUY", "SELL"], help="Order side")
    parser.add_argument("--type", required=True, choices=["MARKET", "LIMIT"], help="Order type")
    parser.add_argument("--quantity", required=True, type=float, help="Quantity to trade")
    parser.add_argument("--price", type=float, help="Limit price (required if type=LIMIT)")

    args = parser.parse_args()

    # Input Validation
    if args.type == "LIMIT" and not args.price:
        parser.error("--price is required for LIMIT orders")

    print("\n Fetching Binance server time...")
    print(get_server_time())

    print("\n Fetching account info...")
    print(json.dumps(get_account_info(), indent=2))

    print("\n Placing order...")
    place_order(args.symbol, args.side, args.type, args.quantity, args.price)


#  ENTRY POINT

if __name__ == "__main__":
    # For Colab/Jupyter support, emulate CLI
    import sys
    if "google.colab" in sys.modules or "ipykernel" in sys.modules:
        sys.argv = [
            "bot.py",
            "--symbol", "BTCUSDT",
            "--side", "BUY",
            "--type", "MARKET",
            "--quantity", "0.001"
        ]
    main()
```

```python

# Binance Futures Simplified Trading Bot with Grid Order Support

# Features:
# - Place Market, Limit, and Grid orders on Binance Futures Testnet
# - Command-line input validation
# - Logging of requests/responses/errors
# - Reusable functions for all operations


import time, hmac, hashlib, requests, json, logging


# CONFIGURATION

API_KEY = "a1XLo08ziWLpcMJ85MLVGpDQ3zDwDGEZssYFDPszCX8scJ7pu3gW0KyOcRgMfY2m"
API_SECRET = "XTT2uCJHdrhnnrf0e6pjWyXHiog3mJbsdcMqZthU7dZEnkuPz5G6oNP85ZqeNmtv"
BASE_URL = "https://testnet.binancefuture.com"   # Testnet Futures base URL

# Logging setup
logging.basicConfig(filename="trading_bot.log",
                    level=logging.INFO,
                    format="%(asctime)s - %(levelname)s - %(message)s")


# Utility Functions

def sign(params: dict) -> dict:
    """Generate signed parameters for Binance API."""
    query = '&'.join([f"{k}={v}" for k, v in params.items()])
    signature = hmac.new(API_SECRET.encode(), query.encode(), hashlib.sha256).hexdigest()
    params['signature'] = signature
    return params

def binance_post(endpoint: str, params: dict):
    """Send a signed POST request to Binance Futures API."""
    url = f"{BASE_URL}{endpoint}"
    headers = {"X-MBX-APIKEY": API_KEY}
    signed_params = sign(params)
    try:
        resp = requests.post(url, headers=headers, params=signed_params)
        resp.raise_for_status()
        logging.info(f"Request success: {resp.json()}")
        return resp.json()
    except requests.exceptions.RequestException as e:
        logging.error(f"HTTP Error: {e} | Response: {resp.text if 'resp' in locals() else 'None'}")
        print(" HTTP Error:", e)
        print("Response:", resp.text if 'resp' in locals() else 'None')
        return None


# Order Functions

def place_market_order(symbol, side, quantity):
    params = {
        "symbol": symbol,
        "side": side,
        "type": "MARKET",
        "quantity": quantity,
        "timestamp": int(time.time() * 1000)
    }
    print(f"\n🔹 Placing MARKET {side} order for {quantity} {symbol}")
    result = binance_post("/fapi/v1/order", params)
    if result:
        print(json.dumps(result, indent=2))
    return result


def place_limit_order(symbol, side, quantity, price):
    params = {
        "symbol": symbol,
        "side": side,
        "type": "LIMIT",
        "timeInForce": "GTC",
        "quantity": quantity,
        "price": price,
        "timestamp": int(time.time() * 1000)
    }
    print(f"\n Placing LIMIT {side} order for {quantity} {symbol} @ {price}")
    result = binance_post("/fapi/v1/order", params)
    if result:
        print(json.dumps(result, indent=2))
    return result


def place_grid_order(symbol, side, quantity, start_price, step, levels):
    """Grid order: places multiple limit orders spaced by step size."""
    print(f"\n🔹 Placing GRID {side} orders for {symbol}")
    results = []
    for i in range(levels):
        price = round(start_price + i * step if side == "SELL" else start_price - i * step, 2)
        print(f"  → Level {i+1}: {side} {quantity} @ {price}")
        result = place_limit_order(symbol, side, quantity, price)
        results.append(result)
        time.sleep(0.5)
    return results

# Command-Line Interface

def main():
    print("=== Binance Futures Simplified Trading Bot ===")
    symbol = input("Enter symbol (e.g., BTCUSDT): ").upper()
    side = input("Enter side (BUY/SELL): ").upper()
    order_type = input("Enter order type (MARKET / LIMIT / GRID): ").upper()

    if order_type not in ["MARKET", "LIMIT", "GRID"]:
        print(" Invalid order type. Use MARKET, LIMIT, or GRID.")
        return

    try:
        quantity = float(input("Enter quantity (e.g., 0.001): "))
    except ValueError:
        print(" Invalid quantity.")
        return

    if order_type == "MARKET":
        place_market_order(symbol, side, quantity)

    elif order_type == "LIMIT":
        try:
            price = float(input("Enter limit price: "))
        except ValueError:
            print(" Invalid price.")
            return
        place_limit_order(symbol, side, quantity, price)

    elif order_type == "GRID":
        try:
            start_price = float(input("Enter starting price: "))
            step = float(input("Enter price step between orders: "))
            levels = int(input("Enter number of grid levels: "))
        except ValueError:
            print(" Invalid grid input.")
            return
        place_grid_order(symbol, side, quantity, start_price, step, levels)

    print("\n Trading operation complete. Check 'trading_bot.log' for details.")


# Run script
if __name__ == "__main__":
    main()
```
